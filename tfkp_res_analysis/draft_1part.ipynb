# Разведовательный анализ данных.

```python
import matplotlib.pyplot as plt
import scipy.stats as sps
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# считываем наши данные

data = pd.read_excel('var3.xlsx')

# посмотрим на данные

data
```

```python
# почистим данные от NaN

data = data.dropna()

data
```

```python
data = data.sort_values('Sum')

data
```

```python
data.describe()
```

```python
# размах

data['Sum'].max() - data['Sum'].min()
```

```python
# мода

data['Sum'].mode()
```

```python
# медиана

data['Sum'].median()
```

```python
# коэффициент ассиметрии

sps.skew(data['Sum'].to_numpy(), bias=False)

# пологость слева
```

## Boxplot

```python
plt.figure(figsize=(13.5, 1))

plt.xlabel('Sum', fontsize=15)

sns.boxplot(data.loc[:, 'Sum'], color='b', saturation=0.5);
```

## ЭДФ и плотность.

```python
original_sums = data['Sum'].to_numpy()

original_sums
```

```python
freqs = data['Sum'].value_counts(normalize=True, sort=False)
freqs = freqs.to_numpy()

freqs
```

```python
probs = np.cumsum(freqs)

probs
```

```python
unique_data = data.drop_duplicates('Sum')
sums = unique_data['Sum'].to_numpy()

probs = np.append(probs, 1)
sums = np.append(sums, 18)
```

```python

plt.figure(figsize=(12.4, 9.6))

plt.step(sums, probs, color='b', where='post');
plt.xlim(sums.min(), sums.max())
plt.ylim(0)

plt.grid()
```
